1. Why Preprocessing?

Your raw data:

Has categorical values (weather, lighting…)
May have missing values
May have inconsistent formats

👉 Preprocessing converts it into machine-readable format

```python
import pandas as pd

df=pd.read_csv("newdataset/train.csv")
```

```python
df.head()
```

. Convert Categorical → Numerical

```python
from sklearn.preprocessing import LabelEncoder

le_dict = {}

for col in df.select_dtypes(include='object').columns:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    le_dict[col] = le   # save encoder for test data
```

```python
x=df.drop("accident_risk",axis=1)
y=df["accident_risk"]
```

🔀 8. Train-Test Split

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42)
```

save processing with pickle otherwise have to retrain model

```python
import pickle
import os

os.makedirs("model")

```

```python
pickle.dump(le_dict,open("model/preprocessor.pkl","wb"))
```
